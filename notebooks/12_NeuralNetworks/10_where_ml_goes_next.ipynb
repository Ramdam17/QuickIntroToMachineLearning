## Part I — what you built

You started with the simplest idea that could possibly work: to classify a point, compare it to the
average of each class and pick the nearest. From there, one method at a time, you built your way up to
networks that *learn* their own features. And at every step you built the idea **by hand first** — the
distance, the probability, the gradient, the split, the margin, the backward pass — and only then met
the library that does it for you. That is why you own the machinery, not merely the imports.

Here is the whole journey in one picture.

```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch

from ml_course import viz
from ml_course.colors import COLORS

viz.use_course_style()

# The twelve methods, grouped into the seven families, in the order you built them.
# Colours are used here as categorical family identities (all from the charter palette).
families = [
    ("Instance-based", ["k-NN"], COLORS["class_a"]),
    ("Probabilistic", ["Naive Bayes"], COLORS["class_b"]),
    ("Linear", ["Logistic Regression"], COLORS["class_c"]),
    ("Partitions", ["Decision Tree"], COLORS["class_d"]),
    ("Margins", ["SVM"], COLORS["class_e"]),
    ("Ensembles", ["Random Forest", "AdaBoost", "Gradient Boosting", "XGBoost", "LightGBM"],
     COLORS["correct"]),
    ("Learned representations", ["MLP", "Neural Networks"], COLORS["error"]),
]

fig, ax = plt.subplots(figsize=(14, 4.2))
n = len(families)
for i, (family, methods, color) in enumerate(families):
    x = i
    # Family box.
    ax.text(x, 1.0, family, ha="center", va="center", fontsize=11, fontweight="bold",
            color=COLORS["text"],
            bbox=dict(boxstyle="round,pad=0.4", facecolor=color, edgecolor=COLORS["text"],
                      linewidth=1.0, alpha=0.85))
    # Methods listed under the family box.
    ax.text(x, 0.42, "\n".join(methods), ha="center", va="top", fontsize=9,
            color=COLORS["text"])
    # Arrow to the next family (the pedagogical spine).
    if i < n - 1:
        ax.add_patch(FancyArrowPatch((x + 0.30, 1.0), (x + 0.70, 1.0),
                                     arrowstyle="-|>", mutation_scale=14,
                                     color=COLORS["muted"], linewidth=1.4))

ax.set_xlim(-0.6, n - 0.4)
ax.set_ylim(-0.35, 1.5)
ax.axis("off")
ax.set_title("The course, end to end: twelve methods, seven families", fontsize=13)
fig.tight_layout()
plt.show()
```

**Read the figure.** Read it left to right — that is the order you built it, and each family is
a different answer to the same question.

- **Instance-based — k-NN.** No training at all: to classify a point, look at the closest stored
  examples. Distance is everything.
- **Probabilistic — Naive Bayes.** Turn each feature into evidence and multiply: class probabilities
  under a bold "features are independent given the class" assumption that works better than it has any
  right to.
- **Linear — Logistic Regression.** A weighted sum of the features, squashed into a calibrated
  probability. The weights are readable evidence, and you trained them with gradient descent.
- **Partitions — Decision Tree.** Axis-aligned if/then rules you can read aloud — and the building
  block of everything in the next family.
- **Margins — SVM.** Not *any* boundary but the one with the widest margin; kernels bend it into
  curves.
- **Ensembles — Random Forest → AdaBoost → Gradient Boosting → XGBoost → LightGBM.** Many imperfect
  models combined: average decorrelated trees to cut variance (bagging), or add weak learners that
  each fix the current mistakes (boosting), then make it regularized and fast.
- **Learned representations — MLP → Neural Networks.** Stop hand-crafting features; let hidden layers
  *learn* them, and build the discipline (initialization, normalization, dropout) to train networks
  deep enough to matter.

## The one idea underneath all of it

Twelve methods look like twelve unrelated tricks until you see what they share. Every one of them is a
different answer to a single question:

> **How do you turn labelled examples into a decision boundary?**

With distances, with probabilities, with a straight line, with rules, with a maximum-margin gap, with
a committee of models, or with a learned representation. That is the whole subject in one sentence —
and now you have built seven different answers to it.

This is not an opinion you were told; it is a result you **measured**, honestly, more than once:

## The through-line you earned #2 — honest evaluation is the real skill

Algorithms come and go; the way you *judge* them is what transfers. Across the course you built a set
of reflexes that apply to **every** model, including ones this course never mentioned:

- always split off a **test set**, and touch it once;
- set a **baseline** (majority class, a linear model) before celebrating any score;
- use **cross-validation**, and report a number **with its spread**, not a lonely decimal;
- keep a **noise ruler** — a difference smaller than the run-to-run wobble is a tie, not a win;
- read the **confusion matrix, the error gallery, the loss curve** — a single accuracy hides where a
  model succeeds and fails.

If you remember nothing else from these thirteen modules, remember this: the score is only as
trustworthy as the evaluation behind it.

## Part II — where machine learning goes next

The last notebook left us on a cliffhanger. Our deep network flattened each 28×28 image into a row of
784 numbers and, in doing so, threw away which pixels were neighbours. Every model we compared did the
same — and a boosted tree, equally blind to the grid, edged it out. The accuracy we left on the table
was not about model *size*; it was about the *structure* none of them used.

That gap points straight at the architectures that run the modern frontier. Three of them, **named and
motivated — not built**. (Building them is the next course, not this one.)

## Convolutional networks (CNNs) — for grids

The direct fix for the flattening problem. Instead of wiring every pixel to every neuron, a CNN slides
a small **filter** across the image and applies the *same* weights everywhere. Two ideas come for
free: **locality** (a filter looks at a small neighbourhood, so adjacency is preserved) and **weight
sharing** (one filter, reused across the whole image, so far fewer parameters and a feature detected in
one corner is detected everywhere). That is the right **inductive bias** for anything on a grid —
photographs, audio spectrograms, board positions. It is exactly what our dense network lacked. (LeCun
et al., 1998.)

```python
import numpy as np

# A tiny 6x6 "image" with visible spatial structure (a plus sign) — schematic, not data.
grid = np.zeros((6, 6))
grid[2:4, :] = 1.0
grid[:, 2:4] = 1.0

fig, (ax_dense, ax_conv) = plt.subplots(1, 2, figsize=(12, 4.6))

# --- Left: the dense view — flatten the grid, wire it all-to-all -------------
ax_dense.imshow(grid, cmap="gray_r", extent=(0, 6, 0, 6))
ax_dense.set_title("Dense: flatten the grid", fontsize=12)
# The flattened row beneath, and a few lines from grid cells to scattered row slots
# (adjacency scrambled — neighbours are no longer neighbours).
for k in range(6):
    ax_dense.add_patch(plt.Rectangle((k, -1.6), 1, 0.8, facecolor=COLORS["muted"],
                                     edgecolor=COLORS["background"], alpha=0.5))
picks = [((2, 2), 0), ((2, 3), 3), ((0, 3), 5), ((4, 4), 1)]
for (r, c), slot in picks:
    ax_dense.add_patch(FancyArrowPatch((c + 0.5, 6 - r - 0.5), (slot + 0.5, -1.2),
                                       arrowstyle="-", color=COLORS["error"],
                                       linewidth=1.0, alpha=0.7))
ax_dense.text(3, -2.4, "neighbours no longer adjacent — locality lost",
              ha="center", va="top", fontsize=9, color=COLORS["error"])
ax_dense.set_xlim(-0.5, 6.5)
ax_dense.set_ylim(-3.2, 6.5)
ax_dense.axis("off")

# --- Right: the convolutional view — one small filter, slid across ----------
ax_conv.imshow(grid, cmap="gray_r", extent=(0, 6, 0, 6))
ax_conv.set_title("Convolution: slide a shared filter", fontsize=12)
# Filter at position A (solid) and B (dashed) + an arrow: it slides, weights shared.
ax_conv.add_patch(plt.Rectangle((0, 3), 3, 3, fill=False, edgecolor=COLORS["highlight"],
                                linewidth=2.5))
ax_conv.add_patch(plt.Rectangle((3, 0), 3, 3, fill=False, edgecolor=COLORS["highlight"],
                                linewidth=2.5, linestyle="--"))
ax_conv.add_patch(FancyArrowPatch((1.5, 3.2), (4.5, 2.8), arrowstyle="-|>", mutation_scale=14,
                                  color=COLORS["highlight"], linewidth=1.6,
                                  connectionstyle="arc3,rad=-0.3"))
ax_conv.text(3, -1.2, "one small filter, shared as it slides — locality kept",
             ha="center", va="top", fontsize=9, color=COLORS["highlight"])
ax_conv.set_xlim(-0.5, 6.5)
ax_conv.set_ylim(-2.0, 6.5)
ax_conv.axis("off")

fig.suptitle("Why images want a CNN: two ways to read the same grid", fontsize=13)
fig.tight_layout()
plt.show()
```

**Read the figure.** On the left, the dense network flattens the grid into a strip: the two
pixels at the centre of the plus sign, right next to each other in the image, land in unrelated slots
of the row (the coral lines). The network cannot *see* that they were neighbours — it would have to
relearn that from data, if it ever does. On the right, a single small filter (rose) slides across the
grid and re-uses the *same* weights at every position: adjacency is preserved, and one detector covers
the whole image with a handful of parameters. That one change — build in locality and weight sharing —
is most of why computer vision runs on convolutional networks.

## Sequences — RNNs and transformers

Images live on a grid; **text, audio and time series live on a sequence** — the order carries the
meaning. Two architectures extend "learn the representation" to ordered data:

- **Recurrent networks (RNNs, and their sturdier cousin the LSTM)** read a sequence one step at a
  time, carrying a running summary — a memory — of what came before.
- **Transformers** drop the step-by-step recurrence for **attention**: every position looks directly
  at every other and weighs what matters to it. That mechanism scales remarkably well, and it is the
  engine inside today's large language models.

Both are the same bet you already made with neural networks — *let the model learn its own
representation* — carried from vectors and grids to sequences. Named, not built. (Hochreiter &
Schmidhuber, 1997; Vaswani et al., 2017.)

## The frontier is built on what you built

It is easy to assume the models behind modern AI are a different world from a course like this one.
They are not. Strip away the scale and the clever architectures and you find the very things you built
by hand in this chapter: **layers** of simple units, trained by **backpropagation** and **gradient
descent**, learning **representations** from data. A transformer with billions of parameters is
trained by the same loop you met in notebook 7 (and wrote by hand before that) — `forward → loss →
backward → step`.

What is new is the scale and the structure (attention, convolution). The **foundations** are the ones
you now own. That is what makes the frontier learnable from here: you are not starting over, you are
adding to a base you built yourself.
